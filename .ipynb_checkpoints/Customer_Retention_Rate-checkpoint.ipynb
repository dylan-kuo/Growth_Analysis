# Customer Retention Rate

Rolling Retention (also called Cohort Analysis)

Rolling retention is defined as the percentage of returning users measure at a regular interval, typical weekly or monthly(I use it for my demo), also known as cohort. By grouping users based on when they signed up, you can gain insight on how your product/marketing/sales initiatives have impacted retention.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import statsmodels.api as sm
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# load data
url = 'data/online_retail.csv'

def load_data(link=url):
    return pd.read_csv(url, encoding='unicode_escape', parse_dates=['InvoiceDate'], infer_datetime_format=True)

df = load_data()
df.head
```

```python
def add_features(df):
    '''
    performs data some transformation as the preprocessing
    '''
    
    # check if these four columns are the subsets of the data frame
    if not set(['InvoiceDate', 'CustomerID', 'Quantity', 'UnitPrice']) <= set(df.columns):  
        print("You don't have the needed columns.")
        return df
    
    # derive year-month format from Invoice date
    df['InvoiceYM'] = df['InvoiceDate'].apply(lambda x: x.strftime('%Y-%m'))
    
    # calculate revenue
    df['Reveneu'] = df['UnitPrice'] * df['Quantity']
    
    # derive sign-up year-month using the min function on Invoice date
    minyearmonth = df.groupby(['CustomerID'])['InvoiceDate'].min().apply(lambda x: x.strftime('%Y-%m')).to_frame('MinYearMonth')
    
    # merge the derived sign-up infos to the original data frame
    df = df.join(minyearmonth, on='CustomerID')
    
    # classify the customer type to new and existing customer by checking if the sign-up date is equal to the invoice date
    df['CustomerType'] = np.where(df['InvoiceYM'] == df['MinYearMonth'], 'New', 'Exist')
    
    return df

retail = add_features(df)
retail.head()
```

```python
def get_year_month(col, format='%Y-%m'):
    '''
    derives the year and month info from the given datetime column
    '''
   
    year = pd.to_datetime(col, format=format).dt.year
    month = pd.to_datetime(col, format=format).dt.month
    return year, month

# derive invoice year and invoice month
inv_year, inv_month = get_year_month(retail['InvoiceYM'])

# derive sign-up year and sign-up month
signup_year, signup_month = get_year_month(retail['MinYearMonth'])

# cohort grouping rule -> (invoice year - signup year) * 12 + (invoice month - sign-up month) + 1
retail['cohortid'] = (inv_year - signup_year) * 12 + (inv_month - signup_month) + 1
```

```python
grouping_count = retail.groupby(['InvoiceYM', 'cohortid'])

# calculate number of unique customer per cohort
cohort_data = grouping_count['CustomerID'].apply(pd.Series.nunique)
cohort_data = cohort_data.reset_index()

# generate pivot table
cohort_counts = cohort_data.pivot(index='InvoiceYM',
                                  columns='cohortid',
                                  values='CustomerID')

new_df = pd.DataFrame(index=cohort_counts.index)

# re-format the pivot table for better visiual display 
for i in range(len(cohort_counts)):
    new_df = new_df.join(cohort_counts.iloc[:, i].shift(-i))

new_df
```

```python
# calculate the retention rate (rule: dvide count per cohort by the total count)
retention_rate = new_df.divide(new_df.iloc[:, 0], axis=0).round(2)
retention_rate
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(retention_rate, annot=True)
plt.show()
```

The first cohort is `2010-12`. In the first transaction month, i.e., `id=1`, the retention is 100%. In the second month, denoted by `id=2`, the retention rate drops to 0.38. The retention rate was fluctuating among some of the months, ending at 0.27 on the 13th month. This cohort has the highest average monthly retention rate.

The lowest retention rate is 0.08 from cohort `2011-04`. There are only 8% of this group of customers left at its 9th transaction month.
